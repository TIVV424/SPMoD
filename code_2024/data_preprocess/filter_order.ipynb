```python
import pandas as pd

# clean up and filter the trip data
order = pd.read_csv("..//..//..//Database//NYC_trip//order_260.csv", index_col=0)
order
```

```python
order["call_time"] = pd.to_datetime(order["call_time"])
order["end_time"] = pd.to_datetime(order["end_time"])
order['delta_min']  = order['end_time'] - order['call_time']
order['delta_min'] = order['delta_min'].apply(lambda x: x.total_seconds()/60)
order = order[order['delta_min'] < 120]
order = order[order['delta_min'] > 0]
start_time = pd.to_datetime("2022-06-01 00:00:00 AM")
end_time = pd.to_datetime("2022-07-01 00:00:00 AM")
order = order[(order["call_time"] > start_time) & (order["call_time"] <= end_time)]
order.to_csv("..//..//..//Database//NYC_trip//order_clean_260.csv")
order
```
